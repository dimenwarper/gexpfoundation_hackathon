```python
import pandas as pd

exp_df = pd.read_csv("OmicsExpressionProteinCodingGenesTPMLogp1.csv", index_col=0)
exp_df.head()
```

```python
gene2ensembl = pd.read_csv("gene2ensembl.csv").dropna().set_index("initial_alias")["converted_alias"].to_dict()
selected_columns = [c for c in exp_df.columns if c.split()[0] in gene2ensembl]
exp_df = exp_df[selected_columns]
exp_df.columns = [gene2ensembl[c.split()[0]] for c in exp_df.columns]
exp_df
```
